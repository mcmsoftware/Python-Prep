10) ¿Se puede mejorar el proceso del punto 9? Utilizar las sentencias break y/ó continue para tal fin

```python
tope_rango=30
n = 0
primo = True
while (n < tope_rango):
    for div in range(2, n):
        if (n % div == 0):
            primo = False
            break
    if (primo):
        print(n)
    else:
        primo = True
    n += 1
```

11) En los puntos 9 y 10, se diseño un código que encuentra números primos y además se lo optimizó. ¿Es posible saber en qué medida se optimizó?

```python
ciclos_break = 0
tope_rango=30
n = 0
primo = True
while (n < tope_rango):
    for div in range(2, n):
        ciclos_break += 1
        if (n % div == 0):
            primo = False
    if (primo):
        print(n)
    else:
        primo = True
    n += 1
print('Cantidad de ciclos: ' + str(ciclos_break))    
```

12) Si la cantidad de números que se evalúa es mayor a treinta, esa optimización crece?

```python
ciclos_totales = 0
ciclos_break = 0
tope_rango=300
n = 0
primo = True
while (n < tope_rango):
    ciclos_totales += 1
    for div in range(2, n):
        ciclos_break += 1
        if (n % div == 0):
            primo = False
    if (primo):
        print(n)
    else:
        primo = True
    n += 1
print('Cantidad de ciclos: ' + str(ciclos_break))
print('Se optimizó a un ' + str(ciclos_break/ciclos_totales) + '% de ciclos aplicando break')
```

13) Aplicando continue, armar un ciclo while que solo imprima los valores divisibles por 12, dentro del rango de números de 100 a 300

```python
ciclo=99
while (ciclo<=300):
    ciclo += 1
    if (ciclo % 12 != 0):
        continue
    print(ciclo, ' es divisible por 12')
```
